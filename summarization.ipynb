```python
import spacy
```

```python
nlp=spacy.load('en_core_web_sm')
```

```python
text="""On May 1, 2025, cities worldwide witnessed large-scale May Day protests, with participants advocating for labor rights and expressing concerns over various political and economic issues. Demonstrations occurred in countries including France, Germany, Italy, Japan, and the United States.
Wikipedia.In the United States, over 1,000 protests were organized under the "50501" movement, uniting unions, student groups, and grassroots organizations. Protesters voiced opposition to the Trump administration's policies, particularly those perceived as undermining labor rights and targeting immigrants. Notably, in Chicago, thousands marched from Union Park to Grant Park, where Mayor Brandon Johnson addressed the crowd, emphasizing the importance of workers' rights and the labor movement.
Wikipedia.Internationally, protests in Tokyo saw anti-war activists gathering outside Mitsubishi Heavy Industries, urging the company to cease involvement in weapons manufacturing. In Istanbul, Turkey, hundreds of protesters were detained during clashes with police. These global demonstrations highlight the enduring significance of May Day as a platform for advocating workers' rights and social justice."""
```

```python
doc=nlp(text)
```

```python

for token in doc:
  print(f'{token.text}:{token.lemma_}--{token.pos_}--{token.dep_}--{token.shape_}--{token.is_alpha}--{token.is_digit}--{token.is_stop}')
```

```python
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python
stopwords=list(STOP_WORDS)
stopwords
```

```python
len(stopwords)
```

```python
doc=nlp(text)
doc
```

```python
tokens=[token.text for token in doc]
tokens
```

```python
word_frequencies={}
for word in doc:
  if word.text.lower() not in stopwords:
    if word.text.lower() not in punctuation:
      if word.text not in word_frequencies.keys():
            word_frequencies[word.text] = 1
      else:
            word_frequencies[word.text] += 1
```

```python
word_frequencies
```

```python
len(word_frequencies)
```

```python
max(word_frequencies.values())
```

```python
for value in word_frequencies:
  word_frequencies[value]=word_frequencies[value]/max(word_frequencies.values())
```

```python
word_frequencies
```

```python
sent_frequencies={}
for sent in doc.sents:
  for word in sent:
    if word.text.lower() in word_frequencies.keys():
      if sent not in sent_frequencies.keys():
        sent_frequencies[sent]=word_frequencies[word.text.lower()]
```

```python
sent_frequencies
```

```python
len(sent_frequencies)
```

```python
max(sent_frequencies.values())
```

```python
from heapq import nlargest
sent_max=len(sent_frequencies)*0.4
```

```python
sent_max
```

```python
summary=nlargest(int(sent_max),sent_frequencies,key=sent_frequencies.get)
```
